```python
#Fixing Null Values
#1) Filling one data for all,
    #syntax :- pandas_datafram.fillna( data )
csv_data.fillna( 5 )    #null is changed to 5
```

```python
#2) Filling all null data with their similar values
    #a) method = bfill :- backward fill
    #b) method = ffill :- forward fill
#csv_data.fillna( method = 'bfill' )
csv_data = csv_data.fillna( method = 'bfill' ) #backwardfill
print( f"csv_data :- \n{ csv_data }\n" )  
```

```python
csv_data.fillna( method = 'bfill', axis = 1) #columns
```

```python
csv_data.fillna( method = 'bfill', axis = 0 ) #rows
```
