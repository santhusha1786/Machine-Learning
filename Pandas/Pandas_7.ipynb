```python
#get_dummies & loc()
import pandas as pd
```

```python
data = pd.read_csv( r"D:\python_materials\ML_Class\Machine_Learning_Codes-master\Excel Data\Social_Network_Ads.csv" )
print(data)
```

```python
#Syntax :-  pd.get_dummies( column_name )
pd.get_dummies( data['Gender'] )
```

```python
#Storing get_dummies value into a variable and then, we will concatinate the dummy_data with our dataset
gender_dummy = pd.get_dummies( data['Gender'] )
print( f"gender_dummy :- \n{ gender_dummy }\n\ntype( gender_dummy ) :- \n{ type( gender_dummy ) }" )
```

```python
data = data.join( gender_dummy )
data
```

```python
#loc
x = data.loc[ :, 'Age' : 'Male' ] # x is assign to age to female (column 2 - 6) 
x
```
