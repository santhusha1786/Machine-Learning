```python
# Three Dimensional Graphs part B
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 1000)
x
```

```python
"""
The axes (an instance of the class plt.Axes) is what we see above: a bounding
box with ticks and labels, which will eventually contain the plot elements that make
up our visualization.
"""
ax = plt.axes()
ax.plot(x, np.sin(x))
plt.show()
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
ax = plt.axes(projection='3d')         #We need to pass this projection as '3d' then, axes will form a 3d Graph
```

```python
xline = np.linspace(0, 15, 1000)
yline = np.sin(xline)
zline = np.cos(xline)
```

```python
xline = 15 * np.random.random(100)
xline
```

```python
yline = np.sin( xline )
yline
```

```python
zline = np.cos( xline )
zline
```

```python
ax = plt.axes( projection = '3d' )
ax.scatter3D(xline, yline, zline, color = 'green', depthshade = True)
plt.show()
```

```python
ax = plt.axes( projection = '3d' )
ax.scatter3D(xline, yline, zline, color = 'green', depthshade = False)
plt.show()
```

```python
ax = plt.axes( projection = '3d' )
ax.scatter3D(xline, yline, zline, color = 'green', depthshade = False, s = 15)   
#s will decide the markersize. By default it's 20.
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
zline = np.arange( 1, 10001 )
zline
```

```python
zline = zline.reshape( 100, 100 )
print(zline.shape)
zline
```

```python
ax = plt.axes(projection='3d')
ax.contour3D(xline, yline, zline, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()

#row of zline must be same as row of zline and zline's column must be same as col of yline.
#means, z.shape = xline_row, yline_row
```

```python
ax = plt.axes(projection='3d')
ax.plot_wireframe(xline, yline, zline, color='black')
ax.set_title('wireframe')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
